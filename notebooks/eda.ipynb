# House Price Prediction - Exploratory Data Analysis
This notebook performs comprehensive EDA on the house price dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load Data

```python
df = pd.read_csv('../data/raw/house_data.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

## 2. Data Overview

```python
df.info()
```

```python
df.describe()
```

```python
print("Missing values:")
df.isnull().sum()
```

## 3. Distribution Analysis

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Feature Distributions', fontsize=16)

for idx, col in enumerate(df.columns):
    ax = axes[idx // 3, idx % 3]
    df[col].hist(bins=30, ax=ax, edgecolor='black')
    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## 4. Correlation Analysis

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
print("Correlation with Price:")
df.corr()['price'].sort_values(ascending=False)
```

## 5. Feature vs Price Analysis

```python
features = ['area', 'bedrooms', 'bathrooms', 'age', 'location_score']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Features vs Price', fontsize=16)

for idx, feature in enumerate(features):
    ax = axes[idx // 3, idx % 3]
    ax.scatter(df[feature], df['price'], alpha=0.5)
    ax.set_xlabel(feature)
    ax.set_ylabel('Price')
    ax.set_title(f'{feature} vs Price')

plt.tight_layout()
plt.show()
```

## 6. Box Plots for Outlier Detection

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Box Plots - Outlier Detection', fontsize=16)

for idx, col in enumerate(df.columns):
    ax = axes[idx // 3, idx % 3]
    df.boxplot(column=col, ax=ax)
    ax.set_title(col)

plt.tight_layout()
plt.show()
```

## 7. Key Insights

```python
print("=" * 50)
print("KEY INSIGHTS")
print("=" * 50)
print(f"\n1. Average house price: ${df['price'].mean():,.2f}")
print(f"2. Price range: ${df['price'].min():,.2f} - ${df['price'].max():,.2f}")
print(f"3. Average area: {df['area'].mean():.2f} sq ft")
print(f"4. Most common bedrooms: {df['bedrooms'].mode()[0]}")
print(f"5. Average house age: {df['age'].mean():.1f} years")
print(f"\nTop 3 features correlated with price:")
print(df.corr()['price'].sort_values(ascending=False)[1:4])
```
